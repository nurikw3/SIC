```python
tokens = text.encode("utf-8")
tokens = list(map(int, tokens))
print(tokens, end=" ")
```

```python
def get_stats(ids: list[int]) -> dict[tuple[int, int], int]:
    counts = {}
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    print(f"info: top_pair: {max(counts.items(), key=lambda x: x[1])}")
    return counts
```

```python
get_stats([10,10,20,1,2,3,10,10,10,3])
```

```python
stats = get_stats(tokens)
top_pair = max(stats, key=stats.get)
top_pair
```

```python
def merge(ids: list[int], pair: tuple[int,int], idx: int) -> list[int]:
    new_ids = []
    i = 0 
    while i < len(ids):
        if i < len(ids) -1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(idx)
            i += 2
        else:
            new_ids.append(ids[i])
            i += 1
    print(f'info {pair} -> {idx}')
    return new_ids
```

```python
def generate_candidates(text: str, max_len=5):
    n = len(text)
    cand = Counter()
    # O (n * l)
    for l in range(1, max_len+1):
        for i in range(n - l + 1):
            substr = text[i:i+l]
            cand[substr] += 1
    return cand
```

```python
def train_unigram(text, vocab_size=50, max_len=5):
    text = "_" + text.replace(" ", "_")
    cand = generate_candidates(text)
    most_common = cand.most_common(vocab_size)
    total = sum(freq for _, freq in most_common)
    vocab = {tok: freq/total for tok, freq in most_common}
    
    tokens_list = list(vocab.keys())
    
    return vocab, tokens_list
```

```python
def encode(text, tokens_list):
    text = "‚ñÅ" + text.replace(" ", "‚ñÅ")
    tokens = []
    i = 0
    while i < len(text):
        matched = None
        for tok in sorted(tokens_list, key=lambda x: -len(x)):
            if text.startswith(tok, i):
                matched = tok
                break
        if matched:
            tokens.append(matched)
            i += len(matched)
        else:
            tokens.append(text[i])
            i += 1
    return tokens
```

```python
def decode(tokens):
    text = "".join(tokens)
    text = text.replace("‚ñÅ", " ")
    return text
```
